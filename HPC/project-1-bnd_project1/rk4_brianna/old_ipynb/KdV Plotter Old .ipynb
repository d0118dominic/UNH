```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.offsetbox import AnchoredText
```

```python
# only need to one once on startup
def plotDat(T0,TN,DT,dtag,stag,v_vals):
        
    # get variables
    M  = int(v_vals[3])
    c  = int(v_vals[4])
    a  = int(v_vals[5])
    L0 = int(v_vals[6])
    LN = int(v_vals[7])
    dx = float(v_vals[8])
    t0 = float(v_vals[9])
    tn = float(v_vals[10])
    dt = float(v_vals[11])
    
    print("c,a = "+str(c)+","+str(a)+"\t"+"x: ["+str(L0)+", "+str(LN)+"], dx = "+str(dx)+"\t "+"t: ["+str(t0)+", "+str(tn)+"], dt = "+str(dt))
    
    for i in range(T0,TN,DT):

        time = round((i-1)*dt,3)

        datf = np.loadtxt(dtag,usecols=(0,i), skiprows=0)
        solf = np.loadtxt(stag,usecols=(0,i), skiprows=0)
        xvals = solf[:,0]
        dvals = datf[:,1]
        svals = solf[:,1]

        plt.plot(xvals, dvals, label="DAT at t = "+str(time))
        plt.plot(xvals, svals,linestyle='dotted',label="SOL at t = "+str(time))
        plt.label='Inline label'

        plt.xlabel("$x$")
        plt.ylabel("$u(x,t)$")
        plt.legend()
        

        plt.legend(title='Key', bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python

# location of data
floc = "data/"

# open latest vars file
varfile = open("data/kdvVARS.txt",'r') 

# read
vlines  = varfile.readlines()
v_name  = []
v_vals  = []
for x in vlines:
    v_name.append(x.split()[0])
    v_vals.append(x.split()[1])
    
# close file
varfile.close()

# get location of DAT and SOL files
dtag = floc+"kdvDAT"+v_vals[1]
stag = floc+"kdvSOL"+v_vals[1]

#print(dtag[1][0])

plotDat(1,75,15,dtag,stag,v_vals)
```
